```python
import pandas as pd
import numpy as np
```

```python
csv= pd.read_csv('dataFile.txt',names=['ID','Name','Designation','ManagerId','Date of Joining','Salary','Commission','Department Number'])
df=pd.DataFrame(csv)
```

# Employee Table

```python
df
```

# Employees in Department Number = 10

```python
df.loc[df['Department Number']==10]
```

# Number of employees in each department

```python
groupedByDept = df.groupby('Department Number')
for x in groupedByDept:
    print("Number of employees in department number: ",x[0],"=",len(x[1]))
```

# Employee with the maximum salary

```python
maxSal=np.max(df['Salary'])
df.loc[df['Salary']==maxSal]
```

# Number of employees in each designation

```python
groupedByDesig = df.groupby('Designation')
for x in groupedByDesig:
    print("Number of",x[0],"=",len(x[1]))
```
